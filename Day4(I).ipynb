```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=np.array([12,23,34,45,49,56,61,71,74,78])
y=np.array([23,31,38,43,49,56,65,71,79,82])
z=np.array([98,89,86,76,71,68,61,54,49,34])
```

# LINE PLOT

```python
plt.title('Line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.plot(x,y ,marker='o',color='brown',mfc='g',ls='--',ms=7);

```

```python
plt.title('Line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.plot(x,y ,marker='<',color='brown',mfc='g',ls='--',ms=7);
plt.grid()
#marker => <,>,^,o,*,H
```

```python
plt.title('Line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.plot(x,y ,marker='H',color='brown',mfc='g',ls='--',ms=10);
plt.grid()
#marker => <,>,^,o,*,H
```

```python
plt.title('Line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.plot(x,y ,marker='*',color='purple',mfc='b',ls='--',ms=9);
plt.plot(x,z, marker='^',color='yellow',mfc='r',ls=':',ms=12);
plt.grid()
#marker => <,>,^,o,*,H
```

# SUB PLOT

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.title('Line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.plot(x,y ,marker='*',color='pink',mfc='b',ls='--',ms=13);

plt.grid()

plt.subplot(2,2,2)
plt.title('Line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.plot(x,z, marker='^',color='black',mfc='r',ls=':',ms=12);
plt.grid()
```

# BAR GRAPH

```python
students=[56,65,46,51,61]

courses=['Java','Python','R','Kotlin','Dart']
```

```python
plt.bar(courses,students);
plt.title('Bar Graph')
plt.xlabel('Courses')
plt.ylabel('Students')
plt.bar(courses,students,color='pink',edgecolor='black',width=0.7,linewidth=3,alpha=0.8);
```

# SCATTER PLOT

```python
plt.scatter(x,y, marker='*');
plt.grid()
```

# PIE CHART 

```python
plt.title('Pie chart')
plt.pie(students,labels = courses, autopct='%.2f%%',shadow=True, radius=1.2,labeldistance=1.1,colors=['cyan','orange','pink','skyblue','purple'],startangle=90,explode=[0,0,0,0,0.2]);
```

```python


plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.title('Line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.plot(x,y ,marker='*',color='pink',mfc='b',ls='--',ms=13);

plt.grid()

plt.subplot(2,2,2)
plt.title('Scatter plot')
plt.scatter(x,y, marker='*');
plt.grid()


plt.subplot(2,2,3)
plt.bar(courses,students);
plt.title('Bar Graph')
plt.xlabel('Courses')
plt.ylabel('Students')
plt.bar(courses,students,color='pink',edgecolor='black',width=0.7,linewidth=3,alpha=0.8);

plt.subplot(2,2,4)
plt.title('Pie chart')
plt.pie(students,labels = courses, autopct='%.2f%%',shadow=True, radius=1.1,labeldistance=1.1,colors=['cyan','orange','pink','skyblue','purple'],startangle=90,explode=[0,0,0,0,0.2]);

plt.savefig('MY PLOT')
```

## BOX PLOT

```python
marks=[67,63,70,71,72,56,51,21,95]  #used to show extreme high and extreme low values
```

```python
plt.boxplot(marks);
```

```python
q3=np.percentile(marks,75)

q1=np.percentile(marks,25)
```

```python
q3,q1
```

```python
iqr=q3-q1
iqr #interquartile range
```

```python
#lower Bound
lb=q1-1.5*iqr

#Upper Bond
up=q3+1.5*iqr
```

```python
lb,up
```

# Seaborn

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips,hue='time');
plt.show()
```

# Histogram



```python
plt.xlabel('Tip Amount')
plt.ylabel('Frequency')
sns.histplot(tips['tip'],kde=True)

plt.show()
```

```python
plt.title('Outliers in tip')
plt.xlabel('Day')
plt.ylabel('tip')
sns.boxenplot(x='day',y='tip',data=tips);
plt.show()
```

```python
sns.pairplot(tips, hue='smoker');
plt.show()
```
