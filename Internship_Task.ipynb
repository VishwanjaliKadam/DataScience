```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV file into a DataFrame
df = pd.read_csv("diamonds.csv")
```

```python
# 1. Print the total number of entries in the dataset
total_entries = len(df)
print("Total number of entries:", total_entries)
```

```python
# 2. Plot the distribution of diamond prices
plt.figure(figsize=(8, 6))
plt.hist(df['price'], bins=30, color='skyblue', edgecolor='black')
plt.title("Distribution of Diamond Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
# 3. Calculate the average price for each cut and color
average_price_cut_color = df.groupby(['cut', 'color'])['price'].mean()
print("Average price for each cut and color:")
print(average_price_cut_color)
```

```python
# 4. Print the names of diamonds with the highest and lowest prices
highest_price_diamond = df.loc[df['price'].idxmax()]['cut']
lowest_price_diamond = df.loc[df['price'].idxmin()]['cut']
print("Diamond with the highest price:", highest_price_diamond)
print("Diamond with the lowest price:", lowest_price_diamond)
```

```python
# 5. Plot the average price by cut and color
plt.figure(figsize=(10, 6))
average_price_cut_color.unstack().plot(kind='bar')
plt.title("Average Price by Cut and Color")
plt.xlabel("Cut")
plt.ylabel("Average Price")
plt.legend(title='Color')
plt.xticks(rotation=45)
plt.show()
```

```python
# 6. Calculate the correlation between diamond carat weight and price
correlation_carat_price = df['carat'].corr(df['price'])
print("Correlation between carat weight and price:", correlation_carat_price)
```

```python
# 7. Plot the distribution of diamond carat weights
plt.figure(figsize=(8, 6))
plt.hist(df['carat'], bins=30, color='lightgreen', edgecolor='black')
plt.title("Distribution of Diamond Carat Weights")
plt.xlabel("Carat Weight")
plt.ylabel("Frequency")
plt.show()
```

```python
# 8. Print the names of diamonds with the highest and lowest carat weights
highest_carat_diamond = df.loc[df['carat'].idxmax()]['cut']
lowest_carat_diamond = df.loc[df['carat'].idxmin()]['cut']
print("Diamond with the highest carat weight:", highest_carat_diamond)
print("Diamond with the lowest carat weight:", lowest_carat_diamond)
```

```python
# 9. Plot the distribution of diamond clarity levels
plt.figure(figsize=(8, 6))
df['clarity'].value_counts().plot(kind='bar', color='lightcoral')
plt.title("Distribution of Diamond Clarity Levels")
plt.xlabel("Clarity")
plt.ylabel("Frequency")
plt.show()
```

```python
# 10. Calculate the average price by clarity level
average_price_clarity = df.groupby('clarity')['price'].mean()
print("Average price by clarity level:")
print(average_price_clarity)
```

```python
# 11. Print the names of diamonds with the highest and lowest price-per-carat

# Calculate price per carat and add it as a new column
df['price_per_carat'] = df['price'] / df['carat']

# Find the index of the row with the highest and lowest price-per-carat values
highest_ppc_diamond_index = df['price_per_carat'].idxmax()
lowest_ppc_diamond_index = df['price_per_carat'].idxmin()

# Retrieve the names of diamonds with the highest and lowest price-per-carat
highest_ppc_diamond_name = df.loc[highest_ppc_diamond_index, 'cut']
lowest_ppc_diamond_name = df.loc[lowest_ppc_diamond_index, 'cut']

# Print the names of diamonds with the highest and lowest price-per-carat
print("Diamond with the highest price-per-carat:", highest_ppc_diamond_name)
print("Diamond with the lowest price-per-carat:", lowest_ppc_diamond_name)


```

```python
# 12. Plot the distribution of diamond prices by clarity level
plt.figure(figsize=(10, 6))
df.groupby('clarity')['price'].mean().plot(kind='bar', color='lightblue')
plt.title("Average Price by Clarity Level")
plt.xlabel("Clarity")
plt.ylabel("Average Price")
plt.show()
```

```python
# 13. Calculate the median price for each cut and color
median_price_cut_color = df.groupby(['cut', 'color'])['price'].median()
print("Median price for each cut and color:")
print(median_price_cut_color)
```

```python
# 14. Print the names of diamonds with the highest and lowest price-per-carat for each clarity level
highest_ppc_per_clarity = df.groupby('clarity')['price_per_carat'].idxmax().apply(lambda x: df.loc[x]['cut'])
lowest_ppc_per_clarity = df.groupby('clarity')['price_per_carat'].idxmin().apply(lambda x: df.loc[x]['cut'])
print("Diamonds with the highest price-per-carat for each clarity level:")
print(highest_ppc_per_clarity)
print("Diamonds with the lowest price-per-carat for each clarity level:")
print(lowest_ppc_per_clarity)
```

```python
# 15. Plot the average price by cut and clarity
plt.figure(figsize=(10, 6))
df.pivot_table(index='cut', columns='clarity', values='price', aggfunc='mean').plot(kind='bar')

plt.title("Average Price by Cut and Clarity")
plt.xlabel("Cut")
plt.ylabel("Average Price")
plt.legend(title='Clarity')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

plt.show()
```
